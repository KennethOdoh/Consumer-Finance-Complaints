```python
def replace_val(df, series, mapper):
    '''
    Function to replace duplicities across a column.
    How to use:
    df: the dataframe you want to work on
    series: the column of the dataframe that contains the duplicities
    mapper: a dictionary whose keys correspond to the desired value, and the value is the item you want to be replaced.
    You can replace multiple values by passing it as a list to the dictionary value.
    
    '''
    for key, value in mapper.items():
        if type(value) == list:
            for item in value:
                df[series].replace(item, key, inplace=True)
        else:
            df[series].replace(value, key, inplace=True)
```

```python
# Let's replace multiple variations of the same sentence with a starndardized one
# on the Issue column
replace_val(df, 'Issue', Issue_duplicity_mapper)
df['Issue'].unique()
```

### Fix Null values in columns

https://cfpb.github.io/api/ccdb/fields.html
The above link is for the reference for filling the form that generated these responses being analysed and as such provided answers to justify the high figures of null values being:

1. some products do not have sub-products. Example is vehicle loan and that will lead to having some null values in the column.
2. issues depend on both products and sub-products. Hence, different products and sub-products will have difference issues.
3. Sub-issue also depends on the trio of products, sub-products and sub-issue. So that will also lead to having null values.
4. As for customer complaint, it can be assumed that quite a number of customers didnt consent to making their complaints public, hence the hugh figures. Same goes for company's response as these are optional fields.
5. Based on this analysis, it is recommended that the 3 columns(consumer complaint narrative, company public response and tags) with more than 60% of null values be dropped.
6. We forward or backward fill the other null values on the remaining columns
7. Apply  label encoding to convert the columns to numbers
